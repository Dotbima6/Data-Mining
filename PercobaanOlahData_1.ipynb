```python
#import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Read dataset
# Membaca dataset dari file Excel
df = pd.read_excel('dataKasus-1.xlsx')

# Melihat beberapa baris pertama dari dataset
df.head()
```

```python
# Melihat informasi umum tentang dataset
print(df.info())
```

```python
# Fungsi untuk menampilkan tipe data dan unique value
def show_column_info(df, column_name):
    print(f"Info kolom '{column_name}' sebelum transformasi:")
    print(f"Tipe data: {df[column_name].dtype}")
    print(f"Nilai unik: {df[column_name].unique()}")

    # Transformasi kolom
    df[column_name] = df[column_name].str.extract('(\d+)')  # Ekstrak angka dari kolom
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')  # Konversi menjadi numerik

    print(f"\nInfo kolom '{column_name}' setelah transformasi:")
    print(f"Tipe data: {df[column_name].dtype}")
    print(f"Nilai unik: {df[column_name].unique()}")
```

```python
# Menampilkan info sebelum dan sesudah transformasi kolom USIA
show_column_info(df, 'USIA')
```

```python
# Jika perlu, simpan hasil yang sudah ditransformasi ke file baru
df.to_excel('transformed_dataKasus-1.xlsx', index=False)
```

```python
# 1. Mengidentifikasi nilai NaN di setiap kolom
print("Jumlah nilai NaN sebelum data cleansing:")
print(df.isnull().sum())
```

```python
# 2. Mengatasi NaN sesuai jenis kolom
# Untuk kolom numerik (contoh: USIA), ganti NaN dengan median
df['USIA'].fillna(df['USIA'].median(), inplace=True)
```

```python
# Untuk kolom kategorikal (contoh: SOSEK RENDAH), ganti NaN dengan modus (nilai yang paling sering muncul)
df['SOSEK RENDAH'].fillna(df['SOSEK RENDAH'].mode()[0], inplace=True)
```

```python
# 3. Mengidentifikasi nilai NaN setelah pembersihan
print("\nJumlah nilai NaN setelah data cleansing:")
print(df.isnull().sum())
```

```python
# Menghapus baris yang memiliki NaN pada kolom 'JARAK KELAHIRAN'
df_cleaned = df.dropna(subset=['JARAK KELAHIRAN'])

# Mengecek apakah ada NaN yang tersisa di kolom 'JARAK KELAHIRAN'
nan_after_cleaning = df_cleaned['JARAK KELAHIRAN'].isnull().sum()

# Menampilkan jumlah NaN setelah proses penghapusan
print(f"Jumlah NaN setelah penghapusan: {nan_after_cleaning}")
print(f"Bentuk data setelah penghapusan: {df_cleaned.shape}")
```

```python
# Jika Anda ingin hanya menerapkan pada kolom tertentu, contoh:
df['RIW HIPERTENSI'] = df['RIW HIPERTENSI'].replace({'Tidak': 0, 'Ya': 1})

df['RIW PE'] = df['RIW PE'].replace({'Tidak': 0})

df['OBESITAS'] = df['OBESITAS'].replace({'Tidak': 0, 'Ya': 1})

df['RIW DM'] = df['RIW DM'].replace({'Tidak': 0, 'Ya': 1})

df['RIW HIPERTENSI/PE DALAM KELUARGA'] = df['RIW HIPERTENSI/PE DALAM KELUARGA'].replace({'Tidak': 0, 'Ada': 1})
```
